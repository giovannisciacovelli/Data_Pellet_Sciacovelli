```python
# Part I: Get all available geographical entities (will need to drop useless ones)
geo = []
url = 'http://dataservices.imf.org/REST/SDMX_JSON.svc/'
key = 'DataStructure/IFS'
dimension_list = requests.get(f'{url}{key}').json()\
            ['Structure']['KeyFamilies']['KeyFamily']\
            ['Components']['Dimension']

key = f"CodeList/{dimension_list[1]['@codelist']}"
code_list = requests.get(f'{url}{key}').json()\
            ['Structure']['CodeLists']['CodeList']['Code']
for code in code_list:
    geo.append(code['@value'])

# geo
```

## What does the variable code means?
FIGB_PA: Financial, Interest Rates, Government Securities, Government Bonds, Percent per annum
FITB_3M_PA: Financial, Interest Rates, Government Securities, Treasury Bills, 3-month, Percent per annum
FII_3M_PA: Financial, Interest Rates, 3-Month Interbank Interest, Percent per annum


```python
# This loop should retrive data for one country at the time, for all counterparts at once
start_time0 = time.time()
one_big_df = dict()
variabls = ['FIGB_PA'] #, 'FIGBY_SM_PA', 'FIGB_S_PA', 'FII_3M_PA', 'FITB_3M_PA', 'FITB_PA']
# geo=['FR']
for var in variabls:
    dfs = list()
    for country in geo: #geo
        print(country)
        key = 'http://dataservices.imf.org/REST/SDMX_JSON.svc/CompactData/IFS/A.' + country +'.' + var + '.' + '?startPeriod=1995&endPeriod=2020'
        try:
            # Navigate to series in API-returned JSON data
            list_data = (requests.get(key).json()
                    ['CompactData']['DataSet']['Series'])
            other_fields = ['@REF_AREA', '@INDICATOR']

    #         for item in list_data:
            try:
                df = pd.DataFrame(list_data['Obs'])
                for field in other_fields:
                    df[field] = list_data[field]
                dfs.append(df)
            except Exception as ex:
                print('Error parsing list item')
                print(item)
                print(str(ex))

            one_big_df[var] = pd.concat(dfs, axis=0)
        except:
            print(f'Country {country} has no Series attribute, skipped')
        time.sleep(0.5)
    if var == 'FIGB_PA':
        try:
            one_big_df[var] = one_big_df[var].drop(columns=['@OBS_STATUS'])
        except:
            print('Error above')
        one_big_df[var].columns = ['Date', 'Long_rate', 'Country', 'Variable_long']
    else:
        try:
            one_big_df[var] = one_big_df[var].drop(columns=['@OBS_STATUS'])
            one_big_df[var].columns = ['Date', 'Short_rate', 'Country', 'Variable_short']
        except:
            print('Error above')
one_big_df_final = (one_big_df['FIGB_PA'])
# one_big_df_final = pd.merge(one_big_df['FIGB_PA'], one_big_df['FIGBY_SM_PA'], how='outer', on=["Date", "Country"])
# one_big_df_final = pd.merge(one_big_df_final, one_big_df['FIGB_S_PA'], how='outer', on=["Date", "Country"])
# one_big_df_final = pd.merge(one_big_df_final, one_big_df['FII_3M_PA'], how='outer', on=["Date", "Country"])
# one_big_df_final = pd.merge(one_big_df_final, one_big_df['FITB_3M_PA'], how='outer', on=["Date", "Country"])
# one_big_df_final = pd.merge(one_big_df_final, one_big_df['FITB_PA'], how='outer', on=["Date", "Country"])

current_time0 = time.time()
print(f"Finished iterating in: {str(int(current_time0 - start_time0))} seconds")
```
