## What does the variable code means?

### In the case of CPIS

Take I_A_D_L_T_BP6_DV_USD.T.T for example
*   I stands for investment
*   A stands for assets
*   D stands for debt securities
*   L stands for long term
*   T stands for total
*   BP6 is the IMF accounting framework
*   DV stands for derived as opposed to measured
*   USD is the currency
*   T.T. is for total holdings


Now what does I_A_T_T_T_BP6_USD.T.T means? Total investment in foreign assets,  Total Holdings, measured in dollars


Note that the CPIS data is based on the outward convention. The "country" holds financial assets from the "counterparty" country. 

### In the case of CDIS
IIW_BP6_USD stands for:
*   I for investment
*   IW for inward (as opposed to OW=outward)
*   Absence of other letters indicate aggregate, as opposed to 
*   DA for debt instrument assets
*   DL for debt instrument liabilities 



So if we want positions to reflect the same side of the balance for CPIS and CDIS, we will need to be careful about inward/outward. Inward is more reliable according to the __[IMF](https://www.imf.org/~/media/Files/Publications/WP/2017/wp17261.ashx)__

With inward flows, the "country" column receives inward investments from "counterparty" countries. 
That means we need to flip the ordering for CPIS for interpretation

```python
# This loop should retrive data for one country at the time, for all counterparts at once
start_time0 = time.time()
one_big_df = dict()
datasets = ['CPIS', 'CDIS']
for dataset in datasets:
    dfs = list()
    for country in geo:
        print(country)
        if dataset == 'CPIS': 
            key = 'http://dataservices.imf.org/REST/SDMX_JSON.svc/CompactData/CPIS/A.' + country +'.I_A_T_T_T_BP6_USD.T.T'
        elif dataset == 'CDIS':
            key = 'http://dataservices.imf.org/REST/SDMX_JSON.svc/CompactData/CDIS/A.' + country +'.IIW_BP6_USD'

        try:
            # Navigate to series in API-returned JSON data
            list_data = (requests.get(f'{key}').json()
                    ['CompactData']['DataSet']['Series'])
            other_fields = ['@REF_AREA', '@COUNTERPART_AREA', '@INDICATOR']

            for item in list_data[:]:
                try:
                    df = pd.DataFrame(item['Obs'])
                    for field in other_fields:
                        df[field] = item[field]
                    dfs.append(df)
                except Exception as ex:
                    print('Error parsing list item')
                    print(item)
                    print(str(ex))

            one_big_df[dataset] = pd.concat(dfs, axis=0)
        except:
            print(f'Country {country} has no Series attribute, skipped')
        time.sleep(0.5)
    one_big_df[dataset] = one_big_df[dataset].drop(columns=['@OBS_STATUS'])
    if dataset == 'CPIS':
        one_big_df[dataset].columns = ['Date', 'Investment_' + dataset, 'Country', 'Counterparty', 'Variable_' + dataset]
    elif dataset == 'CDIS':
        one_big_df[dataset].columns = ['Date', 'Investment_' + dataset, 'Counterparty', 'Country', 'Variable_' + dataset]

one_big_df_final = pd.merge(one_big_df['CPIS'], one_big_df['CDIS'], how='outer', on=["Date", "Country", "Counterparty"])
current_time0 = time.time()
print(f"Finished iterating in: {str(int(current_time0 - start_time0))} seconds")
```
